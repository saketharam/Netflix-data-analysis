### Handling Missing Values

This code fills missing values in specific columns ('rating,' 'cast,' 'country,' 'director') of the DataFrame 'df' with the value 'Unavailable' and modifies the DataFrame in place. It then checks and confirms that there are no remaining missing values in the DataFrame.

```python
X = df["rating"].values
y = df["type"].values
```

This code assigns the "rating" column from the DataFrame 'df' to 'X' and the "type" column to 'y.' Then, it splits the data into training and testing sets, with 70% for training and 30% for testing, using a random state of 41 for reproducibility.

```python
from sklearn.model_selection import train_test_split

X, y = df['rating'], df['type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 41)
```

```python
X_train
```

```python
# Transforming the data using TF-IDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

vectoriser = TfidfVectorizer(ngram_range=(1,2), max_features= 5000)
vectoriser.fit(X_train)
```

```python
X_train = vectoriser.transform(X_train)
X_test = vectoriser.transform(X_test)
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Importing Libraries for the model ,Tfidf and Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline

# define the Classifier
classifier = Pipeline([("tfidf",TfidfVectorizer() ),("classifier",RandomForestClassifier(n_estimators=10))])
```

```python
# Trian Our model
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
# Importing classification_report,accuracy_score,confusion_matrix
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
#confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
#classification_report
classification_report(y_test,y_pred)
```
